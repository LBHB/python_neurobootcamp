## Installing Python with Anaconda

The Anaconda Python distribution is a great way to download and install Python since it also includes many other programs that are popular in scientific computing, such as Jupyter Notebooks, a Python IDE called Spyder, and many of the common Python packages used for scientific computing, among others (a full description of all the packages that are included can be found here https://www.anaconda.com/distribution/).

To get started, download the **Python 3.6** version of Anaconda for your computer via this link: https://www.anaconda.com/download/

## Testing your installation
After you download Ananconda, make sure you open it and get a little familar with it! Here are instructions for opening Jupyter Notebooks on Windows, Mab, and Linux machines. You can open Jupyter Notebook directly or through the Anaconda Navigator. We will go through both.  

### Windows

**Option 1.** From Anaconda Navigator
> From the start menu, type “anaconda navigator” in the search bar and open the Anaconda Navigator application. 
<img src="img/anacondaNavWindows.png"  width="250"/> 
This will open the Ananconda Navigator, where you can explore the different packages that came with anaconda (such as the IDE Spyder and Jupyter Notebook). 

**Option 2.** Open Jupyter Notebook directly
> From the start menu, type “Jupyter Notebook” in the search bar and open the Jupyter Notebook application. 
<img src="img/jupyterWindows.png", width = 250 />
Two windows will open: a command prompt and web browser. 


The command prompt will look something like this: 
<img src="img/jupyter3.png"width="500"/>  
And the webrowser should look like this:
<img src="img/jupyter4.png" width="700"/>

The web browser will automatically be populated with the files in your current directory. 

Now, using Jupyter Notebook's user interface, navigate to the Jupyter Notebook (.ipynb file) you want to open and click on it. The Notebook will open in a new tab. 


### Mac

**Option 1.** Find the anaconda navigator:  
> A. From a Finder window:  
    Search for “anaconda-navigator” on your mac. Double click the “anaconda-navigator” icon to launch the anaconda navigator.  
> B. From a terminal window:  
    Open a terminal window and type `anaconda-navigator`  
> Explore and launch different packages that came with anaconda. 


**Option 2.** To directly acces Jupyter Notebooks:
> A. From a Finder window:
    Search for “Jupyter” on your mac. Double click the “jupyter-notebook” icon to launch Jupyter notebook. 
> B. From a terminal window:
    Type `jupyter notebook`. If that doesn't work, try finding anaconda first and opening it from there:  
    The default Anaconda install location is in your home directory. Open a new terminal window and type `ls` into the terminal to list all the files in your home directory, you shoud see it listed. Then execute the following commands:  
> `cd anaconda`  
> `cd bin`  
> `./jupyter-notebook`  

This screenshot below shows the commands mentioned above to open the Jupyter-Notebook application.

<img src="img/jupyterMac.png"/>

The web browser will automatically be populated with the files in your c:\users\your_username\ directory (or the directory you opened Jupyter Notebook from). 

Now, using Jupyter Notebook's user interface, navigate to the Jupyter Notebook (.ipynb file) you want to open and click on it. The Notebook will open in a new tab. 

### Linux: 
    Open a new terminal. Try typing `anaconda-navigator` to open the Ananconda Navigator. If that doesn't work, to navigate toward your directory that contains Anaconda. Navigate to the bin folder. Run the anaconda-navigator or jupyter-notebook executable that is included with Anaconda.
